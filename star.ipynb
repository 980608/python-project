```python
# 왜 우리동네에는 스타벅스가 없을까!!!

# 서울 전 지역 약 500개가 넘은 스타벅스 매장 이 있다. 그렇다면 어느 구에
# 주로 더 많이 입점을 해 있을까. 인구가 많은 구에 스타벅스가 많을까, 아니면
# 어떤 다른 입지 전략을 가지고 있을까

# 두가지 가설...
# 1. 거주 인구가 많은 지역에 스타벅스가 많이 있을 것이다.
# 2. 직장인이 많은 지역에 스타벅스가 많이 있을 것이다.

# 먼저 입지 전략을 파악하기 위해서는 서울 시내에 스타벅스들의 위치들을 파악해야한다.
# 그래서 위에서 제시한 두 가지 가설을 검정하기 위해 인구 통계 데이터 수집도 필요하다.
```

```python
from selenium import webdriver      # 인터넷 연결(크롬)
from bs4 import BeautifulSoup as bs # 분석하기 쉽게 가공(인터넷에서 코드가져오기)
import pandas as pd                 # 데이터 분석
```

```python
driver = webdriver.Chrome('chromedriver.exe')
url = 'https://www.starbucks.co.kr/store/store_map.do' 
driver.get(url)
```

```python
# 지역 검색 단추 클릭하기

area_btn = '#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a'
driver.find_element_by_css_selector(area_btn).click()
```

```python
# 서울 단추 클릭
seoul_btn = '#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a'
driver.find_element_by_css_selector(seoul_btn).click()
```

```python
# 전체 단추 클릭
all_btn = '#mCSB_2_container > ul > li:nth-child(1) > a'
driver.find_element_by_css_selector(all_btn).click()
```

```python
html = driver.page_source
soup = bs(html,'html.parser') # 문자에서 코드로 변경하여 불러오기
soup
```

```python
starbucks_listall = soup.select('li.quickResultLstCon')
starbucks_listall[0]
```

```python
starbucks_store = starbucks_listall[0]
lat = starbucks_store['data-lat'].strip()
long = starbucks_store['data-long'].strip()
# name = starbucks_store['data-name']
print(lat,long,sep=' / ')
```

```python
starbucks_list = []

for item in starbucks_listall:
    lat = item['data-lat'].strip()
    long = item['data-long'].strip()
    name = item('strong')[0].text.strip()
    address = str(item.select('p.result_details')[0]).split('<br/>')[0].split('>')[1]
    tel = str(item.select('p.result_details')[0]).split('<br/>')[1].split('<')[0]
    starbucks_list.append([name, address, tel, lat, long])
    
starbucks_list
```

```python
columns = ['매장명', '주소', '전화번호', '위도', '경도']
seoul_starbucks_df = pd.DataFrame(starbucks_list, columns = columns)
seoul_starbucks_df
```

```python
sgg_names = []

for address in seoul_starbucks['주소']:
    sgg = address.split()[1]
    sgg_names.append(sgg)
    
sgg_names
    
```

```python
seoul_starbucks['시군구명'] = sgg_names
seoul_starbucks.head()
```

```python
starbucks_sgg_count = seoul_starbucks.pivot_table(index = '시군구명',
                                                  values = '매장명',
                                                  aggfunc = 'count')
starbucks_sgg_count
```

```python
starbucks_sgg_count = seoul_starbucks.pivot_table(index = '시군구명',
                                                  values = '매장명',
                                                  aggfunc = 'count'
                                                 ).rename(columns = {'매장명' : '스타벅스_매장수'})
starbucks_sgg_count
```

```python
seoul_sgg = pd.read_excel('sgg_pop.xlsx')
seoul_sgg
```

```python
seoul_sgg = pd.merge(seoul_sgg,starbucks_sgg_count, how = 'left', on = '시군구명')

seoul_sgg.head()
```
